7. Есть ли связь между зарплатой и опытом, возрастом и опытом, полом и опытом, состоянием в браке и опытом?

**Поскольку большинство данных, представленных в DataFrame, являются строковыми типами данных, что вы можете сделать, чтобы рассчитать матрицу корреляции и увидеть взаимосвязь между столбцами?**

Пожалуйста, работайте с копией pd_data, так как вам понадобится pd_data в следующих упражнениях.

````python
test_data=pd_data.copy()
print(test_data.head())
# Начало вашего кода
test_data['experience'] = test_data['experience'].astype('category').cat.codes
test_data['gender'] = test_data['gender'].astype('category').cat.codes
test_data['married'] = test_data['married'].astype('category').cat.codes
print(test_data['salary'].corr(test_data['experience']))
print(test_data['age'].corr(test_data['experience']))
print(test_data['gender'].corr(test_data['experience']))
print(test_data['married'].corr(test_data['experience']))
# Конец вашего кода
````

````python
'''
Ваши наблюдения: Наиболее высокая корреляция наблюдается между опытом и зарплатой,
                 нежели чем между опытом и остальными категориями данных. Согласно этим наблюдениям, можно сделать вывод,
                 что зарплата и опыт взаимосвязаны между собой и влияют друг на друга. Чем больше опыта, тем выше зарплата.
''';
````
